```python
import requests
import time
from bs4 import BeautifulSoup
```

```python
#Fetch webpage
url = "https://www.rev.com/blog/transcripts/elon-musk-interview-with-don-lemon"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
time.sleep(2)
response = requests.get(url, headers=headers)
```

```python
#Parse it
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
for p in soup.find_all('p'):
    print(p.text)
    print('---')  # Separator between paragraphs
```

```python
elon_mentions = soup.find_all(text=lambda text: 'Elon Musk:' in text if text else False)
for mention in elon_mentions:
    print(mention.parent.text)
    print('---')
```
