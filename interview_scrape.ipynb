Elon interview with Trump
URL: https://turboscribe.ai/transcript/share/4422534834081521519/HWE18owsC2u8E5u2HpZNikyBdermlV2YSwGlTEPKJJw/donald-trump-and-elon-musk-full-transcript-august-12-2024-https-x-com-i-spaces-1nakepnklwoxl

```python
import re

def extract_elon_musk_speech(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Split the content into sections
    sections = re.split(r'\n(?=Elon Musk|Donald Trump)', content)

    # Extract Elon Musk's speech
    elon_musk_speech = []
    for section in sections:
        if section.startswith('Elon Musk'):
            # Remove the "Elon Musk" header and any leading/trailing whitespace
            speech = section.replace('Elon Musk', '', 1).strip()
            elon_musk_speech.append(speech)

    return elon_musk_speech

# Specify the file path
file_path = 'data/elon_trump.txt'

# Extract Elon's speech
elon_speech = extract_elon_musk_speech(file_path)

# Print the extracted speech
for i, speech in enumerate(elon_speech, 1):
    print(f"{speech}")
    print("-" * 50)  # Add a separator between speeches for better readability
```

```python
#Set up list which contains all speeches
speech_total = []

#Append elon_speech to speech_total
speech_total.extend(elon_speech)
```

Elon interview with Don Lemon URL: https://www.rev.com/blog/transcripts/elon-musk-interview-with-don-lemon

```python
def extract_elon_musk_speech(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Split the content into lines
    lines = content.split('\n')

    elon_musk_speech = []
    current_speaker = ""
    current_speech = ""

    for line in lines:
        # Check if the line starts with a name and timestamp
        match = re.match(r'(.+?)\s*\((\d+:\d+)\):', line)
        if match:
            # If we were previously capturing Elon's speech, save it
            if current_speaker == "Elon Musk" and current_speech:
                elon_musk_speech.append(current_speech.strip())
                current_speech = ""
            
            current_speaker = match.group(1)
        elif current_speaker == "Elon Musk":
            # If the current speaker is Elon, add this line to the current speech
            current_speech += line + " "

    # Add the last speech if it's Elon's
    if current_speaker == "Elon Musk" and current_speech:
        elon_musk_speech.append(current_speech.strip())

    return elon_musk_speech

# Specify the file path
file_path = 'data/elon_donlemon.txt'

# Extract Elon's speech
elon_speech = extract_elon_musk_speech(file_path)

# Print the extracted speech
for i, speech in enumerate(elon_speech, 1):
    print(f"{speech}")
    print("-" * 50)  # Add a separator between speeches for better readability

#Append to speech_total
speech_total.extend(elon_speech)
```

Elon interview with Watternberg URL: https://www.pbs.org/thinktank/transcript1292.html

```python
def extract_elon_musk_speech(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    
    lines = content.split('\n')
    elon_musk_speech = []
    current_speech = ""
    is_elon_speaking = False
    
    for line in lines:
        if line.startswith("MUSK:"):
            if current_speech:
                elon_musk_speech.append(current_speech.strip())
            current_speech = line.replace("MUSK:", "", 1).strip()
            is_elon_speaking = True
        elif line.startswith("WATTENBERG:") or any(name.strip() + ":" in line for name in ["WATTENBERG", "THINK TANK"]):
            if current_speech:
                elon_musk_speech.append(current_speech.strip())
                current_speech = ""
            is_elon_speaking = False
        elif is_elon_speaking:
            current_speech += " " + line.strip()
    
    if current_speech:
        elon_musk_speech.append(current_speech.strip())
    
    return elon_musk_speech

# Usage
file_path = 'data/elon_wattenberg.txt'
elon_speech = extract_elon_musk_speech(file_path)

# Print the extracted speech
if elon_speech:
    for speech in elon_speech:
        print(f"{speech}\n")
        print("-" * 50)  # Add a separator between speeches for better readability


speech_total.extend(elon_speech)
```

```python
def extract_elon_musk_speech(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Regular expression to match Elon Musk's speech parts
    pattern = r'Elon Musk\n\([0-9:]+\)(.*?)(?=\n[A-Za-z ]+\n\([0-9:]+\)|\Z)'
    
    # Find all matches
    matches = re.findall(pattern, content, re.DOTALL)
    
    # Clean up the extracted text
    cleaned_speeches = [speech.strip() for speech in matches]
    
    return cleaned_speeches

# File path
file_path = 'data/elon_lex.txt'

# Extract speeches
elon_speech = extract_elon_musk_speech(file_path)

# Print results
for i, speech in enumerate(elon_speech, 1):
    print(speech)
    print("-" * 50)

#Append elon_speech to speech_total
speech_total.extend(elon_speech)
```

Elon interview with Ford URL: https://www.thehenryford.org/documents/default-source/default-document-library/transcript_musk_full-length.pdf?sfvrsn=f5722f01_0

```python
def extract_elon_quotes(file_path, speech_list):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    
    # Regular expression to match Elon's quotes
    pattern = r'ELON MUSK:\s*\d{2}:\d{2}:\d{2};?\d{2}\s*(.*?)(?=(?:BARRY HURD:|ELON MUSK:|\Z))'
    
    # Find all matches
    matches = re.findall(pattern, content, re.DOTALL)
    
    # Process and add each quote to the list
    for quote in matches:
        # Remove page numbers and timestamps
        cleaned_quote = re.sub(r'PG\.\d+|^\d{2}:\d{2}:\d{2};?\d{2}', '', quote, flags=re.MULTILINE)
        
        # Remove any remaining timestamps (in case they're in the middle of text)
        cleaned_quote = re.sub(r'\d{2}:\d{2}:\d{2};?\d{2}', '', cleaned_quote)
        
        # Clean up whitespace
        cleaned_quote = re.sub(r'\s+', ' ', cleaned_quote).strip()
        
        if cleaned_quote:
            speech_list.append(cleaned_quote)

# Use the function
file_path = 'data/elon_ford.txt'  # Make sure this matches your file name
extract_elon_quotes(file_path, speech_total)
```

Elon interview with Faber URL: cnbc-exclusive-cnbc-transcript-elon-musk-sits-down-with-cnbcs-david-faber-live-on-cnbc-tonight-.html

```python
def extract_elon_quotes(file_path, speech_list):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    
    # Use regex to find all of Elon's quotes
    pattern = r'MUSK:(.*?)(?=\n\n(?:DAVID FABER|FABER):|\Z)'
    quotes = re.findall(pattern, content, re.DOTALL)
    
    # Add each quote to the speech_list
    for quote in quotes:
        # Remove any leading/trailing whitespace and newlines
        cleaned_quote = ' '.join(quote.strip().split())
        if cleaned_quote:  # Only add non-empty quotes
            speech_list.append(cleaned_quote)

    return len(quotes)  # Return the number of quotes added

# Use the function
file_path = 'data/elon_faber.txt'
quotes_added = extract_elon_quotes(file_path, speech_total)
```

Elon interview with Ted URL: https://transcript.lol/read/tedtalk/elon-musk/64a0b778503edabcbda8a4d0

```python
def extract_speaker2_quotes(file_path, speech_list):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    
    # Use regex to find all of Speaker 2's quotes
    pattern = r'Speaker 2\n\n(\d{2}:\d{2})\n\n(.*?)(?=\n\nSpeaker \d|\Z)'
    quotes = re.findall(pattern, content, re.DOTALL)
    
    # Add each quote to the speech_list
    for timestamp, quote in quotes:
        # Remove any leading/trailing whitespace and newlines
        cleaned_quote = ' '.join(quote.strip().split())
        if cleaned_quote:  # Only add non-empty quotes
            speech_list.append(f"[{timestamp}] {cleaned_quote}")

    return len(quotes)  # Return the number of quotes added

# Use the function
file_path = 'data/elon_ted.txt'  # Replace with your actual file path
quotes_added = extract_speaker2_quotes(file_path, speech_total)
```

```python
# Add shitelonsays scrape to list

# Define function to add speeches to our list
def add_speeches_to_list(input_file, speech_list):
    with open(input_file, 'r') as infile:
        for line in infile:
            # Remove leading/trailing whitespace and skip empty lines
            line = line.strip()
            if line:
                speech_list.append(line)

# Initalize input_file and run function
input_file = 'data/shitelonsays_scrape.txt'
add_speeches_to_list(input_file, speech_total)
```
