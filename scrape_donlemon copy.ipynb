```python
elon_mentions = soup.find_all(text=lambda text: 'Elon Musk:' in text if text else False)
for mention in elon_mentions:
    print(mention.parent.text)
    print('---')
```

```python
import requests
from bs4 import BeautifulSoup
from lxml import etree
import re

url = "https://lexfridman.com/elon-musk-4-transcript/"
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'lxml')

# Convert to lxml object
dom = etree.HTML(str(soup))

# Use XPath to find the transcript content
transcript_element = dom.xpath("/html/body/table/tbody/tr[2]/td/table/tbody/tr/td[3]/font[2]/h3")

if transcript_element:
    transcript = etree.tostring(transcript_element[0], method="text", encoding="unicode")
    
    # Split the transcript into individual dialogue parts
    dialogue_parts = re.split(r'((?:MUSK|FRIDMAN):)', transcript)
    
    # Combine speaker with their dialogue
    combined_dialogue = [''.join(dialogue_parts[i:i+2]).strip() for i in range(0, len(dialogue_parts), 2) if dialogue_parts[i]]
    
    print("Elon Musk's dialogue:")
    for dialogue in combined_dialogue:
        if dialogue.startswith("MUSK:"):
            print(dialogue)
            print()  # Add a blank line for readability
else:
    print("Transcript content not found using the provided XPath.")
```
